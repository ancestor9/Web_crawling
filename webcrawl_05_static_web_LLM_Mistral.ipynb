```python
# prompt: url 의 page를 1부터 5까지 변경하면서 상품정보를 크롤링해줘

# 기본 URL
BASE_URL = "https://startcoding.pythonanywhere.com/basic"

all_products = []

for page in range(1, 6):
    url = f"{BASE_URL}?page={page}"
    print(f"Crawling page: {url}")
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        products = []
        for product in soup.select('.product'):
            try:
                category = product.select_one('.product-category').text.strip()
                name = product.select_one('.product-name a').text.strip()
                link = product.select_one('.product-name a')['href']
                price = product.select_one('.product-price').text.strip()

                products.append({
                    'category': category,
                    'name': name,
                    'link': link,
                    'price': price
                })
            except AttributeError as e:
                print(f"Error processing product on page {page}: {e}")
                # Consider adding more robust error handling here,
                # such as logging the problematic product HTML
        all_products.extend(products)
    else:
        print(f"Failed to retrieve page {page}: Status code {response.status_code}")

df = pd.DataFrame(all_products)
df
```

```python
# prompt: 예를 들어 df['price'] 값의 '85,900원\n\t\t\t\t\t\t\t\t\'  이 값을 '85900'으로 integer로 변경해줘

import re

def clean_price(price_str):
  """Cleans and converts a price string to an integer.

  Args:
    price_str: The price string to clean.

  Returns:
    An integer representing the cleaned price, or NaN if the price string is invalid.
  """
  try:
    # Remove non-numeric characters, commas, and whitespace
    cleaned_price = re.sub(r"[^0-9]", "", price_str)

    # Convert to integer
    return int(cleaned_price)
  except (ValueError, TypeError):
    return None  # Or handle the error as needed

# Apply the cleaning function to the 'price' column
df['price'] = df['price'].apply(clean_price)
df.head()
```
