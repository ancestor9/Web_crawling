### 1. 서버의 실제 API 엔드포인트 활용 (프론트엔드 우회)
- 많은 동적 웹페이지는 **클라이언트(JavaScript)**에서 데이터를 요청하여 화면을 구성합니다.
- 브라우저 개발자 도구(F12) → Network 탭에서 XHR 또는 Fetch 요청을 확인하면, 서버에서 데이터를 반환하는 실제 API URL을 찾을 수 있습니다.
- 이 API URL을 curl이나 requests를 사용해 직접 호출하면, Selenium을 실행할 필요 없이 데이터를 JSON 형식으로 바로 가져올 수 있습니다.
👉 즉, 프론트엔드 렌더링을 거치지 않고도 데이터를 받아올 수 있음!
👉 HTML 파싱이 필요 없고, 더 정확하고 빠르게 원하는 데이터를 가져올 수 있음!
### 2.  방법
- 2.1. http://land.gangnam.go.kr/land/rtms/rtmsApartment.do 클릭
- 2.2. 강남구 개포동을 검색

- 2.4. 특정 가격을 클릭하여 검색해보면 http://land.gangnam.go.kr/land/rtms/getRtmsPriceList.do 이 나오는 것을 확인
- 2.5. 오른쪽 클릭 copy > Copy as cURL(bash)
- 2.6. https://curlconverter.com/python/ 붙여넣기 하고 출력 python code는 아래 (cookies는 없어도 됨)
- 2.7. tr의 td를 가져오게 해서 데이터프레임으로 ㅁ나들고 데이저 정제 (AI에게)

```python
import requests


cookies = {
    'WL_PCID': '17399251067915276377879',
    '_ga': 'GA1.1.243957099.1739925107',
    '_ga_P3LRGGQYFZ': 'GS1.1.1739927646.2.1.1739927965.0.0.0',
}

headers = {
    'Accept': 'application/json, text/javascript, */*; q=0.01',
    'Accept-Language': 'ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
    'Content-Type': 'application/x-www-form-urlencoded; charset=UTF-8',
    # 'Cookie': 'WL_PCID=17399251067915276377879; _ga=GA1.1.243957099.1739925107; _ga_P3LRGGQYFZ=GS1.1.1739927646.2.1.1739927965.0.0.0',
    'Origin': 'http://land.gangnam.go.kr',
    'Referer': 'http://land.gangnam.go.kr/land/rtms/rtmsApartment.do',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/132.0.0.0 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest',
}

data = '&sggCd=11680&rightGbnGrp=RTMS&bjdong10Cd=10300&selectGubun=1&rnCd=&startDate=20250101&endDate=20250331&bldgNm=&bldgGbn=AP'

response = requests.post(
    'http://land.gangnam.go.kr/land/rtms/getRtmsPriceList.do',
    cookies=cookies,
    headers=headers,
    data=data,
    verify=False,
)

```

```python
print(response.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
tr_list = [tr for tr in soup.find_all('tr')]
print(tr_list)
```

```python
td_list = [td for td in soup.find_all('td')]
print(td_list)
```

```python
import pandas as pd
data = [[td.get_text(strip=True) for td in tr.find_all('td')] for tr in tr_list]
df = pd.DataFrame(data)
print(df)
```
