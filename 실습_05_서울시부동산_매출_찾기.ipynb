```python
print(response.text)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
tr_list = [tr for tr in soup.find_all('tr')]
print(tr_list)
```

```python
td_list = [td for td in soup.find_all('td')]
print(td_list)
```

```python
import pandas as pd
data = [[td.get_text(strip=True) for td in tr.find_all('td')] for tr in tr_list]
df = pd.DataFrame(data)
print(df)
```
